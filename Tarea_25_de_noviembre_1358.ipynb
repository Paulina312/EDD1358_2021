```
class NodoDoble:
    def __init__( self , value , previo = None , siguiente = None ):
        self.data = value
        self.prev = previo
        self.next = siguiente

class DoubleLinkedList:
  def __init__(self):
    self.__head = None
    self.__tail = None
    self.__size = 0

    """
    Ejemplos de como podemos comenzar:
    NodDoble(10)
    NodoDoble(10, NodoDoble(5), NodoDoble(29))
    """

  def get_size(self):
    return self.__size

  def is_empty(self):
    return self.__size == 0

  def append(self, value):
    if self.is_empty():
      nuevo = NodoDoble(value)
      self.__head = nuevo
      self.__tail = nuevo
    else:
      nuevo = NodoDoble(value, self.__tail, None)
      self.__tail.next = nuevo
      self.__tail = nuevo
    self.__size += 1
  
  def find_from_head(self, value):
    if self.is_empty():
      pass
    else:
     curr_node = self.__head
     posicion = 0
     while curr_node.data != value and curr_node != None:
       curr_node = curr_node.next
       posicion = posicion+1
     if curr_node.data == value:
       print(posicion) 
     else:
       print("El nodo no existe")

  def find_from_tail(self, value):
    if self.is_empty():
      pass
    else:
      curr_node = self.__tail
      posicion = self.__size-1
      while curr_node.data != value and curr_node != None:
        curr_node = curr_node.prev
        posicion = posicion-1
      if curr_node.data == value:
        print(posicion)
      else:
        print("El nodo no existe")

  def remove_from_head(self, value):
    if self.is_empty():
      pass
    else:
      curr_node = self.__head
      while curr_node.data != value and curr_node != None:
        curr_node = curr_node.next
      if curr_node.data == value and curr_node == self.__head:
        self.__head = curr_node.next
        curr_node.next.prev = None
        self.__size -= 1
      elif curr_node.data == value:
        curr_node.prev.next = curr_node.next 
        curr_node.next.prev = curr_node.prev
        self.__size -= 1
      else:
        print("El nodo no existe")

  def remove_from_tail(self, value):
    if self.is_empty():
      pass
    else:
      curr_node = self.__tail
      while curr_node.data != value and curr_node != None:
        curr_node = curr_node.prev
      if curr_node.data == value and curr_node == self.__tail:
        self.__tail = curr_node.prev
        curr_node.prev.next = None
        self.__size -= 1
      elif curr_node.data == value:
        curr_node.prev.next = curr_node.next
        curr_node.next.prev = curr_node.prev
        self.__size -=1
      else:
        print("El nodo no existe")
  
  def transversal(self):
    curr_node = self.__head
    while curr_node != None:
      print("<---", curr_node.data, "--->", end= "") # el end="" es para que no haga salto de l√≠nea
      curr_node = curr_node.next
    print("")
  
  def reverse_transversal(self):
    curr_node = self.__tail
    while curr_node != None:
      print("<---", curr_node.data, "--->", end= "")
      curr_node = curr_node.prev
    print("")
```

```
ldl = DoubleLinkedList()
print(ldl.get_size())
ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.append(40)
ldl.append(50)
ldl.append(60)
ldl.append(70)
ldl.transversal()
ldl.find_from_head(30)
ldl.find_from_tail(30)
ldl.remove_from_head(20)
ldl.transversal() 
ldl.remove_from_tail(70)
ldl.transversal()
ldl.reverse_transversal()
```
