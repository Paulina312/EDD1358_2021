```
class NodoArbol:
  def __init__(self, value, left = None, right = None):
    self.data= value
    self.left= left
    self.right= right
```

```
class ArbolBinarioBusqueda:
  def __init__(self):
    self.__root = None

  def insert(self, value):
    # cuidado con mover el valor de root!!! es nuestra referencia, perderíamos la entrada al árbol
    if self.__root == None:
      self.__root = NodoArbol(value)
    else:
      self.__insert__(self.__root, value)
  
  def __insert__(self, nodo, value): # método recursivo
    if nodo.data == value:
      print("El nodo ya existe, no se inserta ", value)
    elif value < nodo.data:
      # ir a la izquierda
      if nodo.left == None:
        nodo.left = NodoArbol(value)
      else: 
        self.__insert__(nodo.left, value)
    else:
      # ir a la derecha
      if nodo.right == None:
        nodo.right = NodoArbol(value)
      else: 
        self.__insert__(nodo.right, value)

  def transversal(self, format="in"):#in, pos, pre
    if format == "in":
      print("\nRecorrido in orden:")
      self.__recorrido_in(self.__root)
    elif format == "pre":
      print("\nRecorrido en pre orden:")
      self.__recorrido_pre(self.__root)
    elif format == "pos":
      print("\nRecorrido en pos orden:")
      self.__recorrido_pos(self.__root)
    else:
      print("Formato no valido")
  
  def __recorrido_in(self, nodo): # método recursivo
    if nodo == None:
      #print("caso base")
      pass
    else:
      self.__recorrido_in(nodo.left)
      print(nodo.data, end= ",")
      self.__recorrido_in(nodo.right)

  def __recorrido_pre(self, nodo): # método recursivo
    if nodo == None:
      #print("caso base")
      pass
    else:
      print(nodo.data, end= ",")
      self.__recorrido_pre(nodo.left)
      self.__recorrido_pre(nodo.right)

  def __recorrido_pos(self, nodo): # método recursivo
    if nodo == None:
      #print("caso base")
      pass
    else:
      self.__recorrido_pos(nodo.left)
      self.__recorrido_pos(nodo.right)
      print(nodo.data, end= ",")
  
  def search(self, value):
    if self.__root == None:
      print("árbol vacio")
    else:
      return self.__search(self.__root, value)
  
  def __search(self, nodo, value): # método recursivo
    if nodo == None:
      return None
    elif nodo.data == value:
      return nodo
    elif value < nodo.data:
      print("ir a la izquierda")
      return self.__search(nodo.left, value)
    else:
      print("ir a la derecha")
      return self.__search(nodo.right, value)
  
  def remove(self, value):
    if self.__root == None:
      print("Árbol vacío")
      #el que deseamos eliminar es la raíz
    else:
      self.__remove(None, None, self.__root, value)

  def __remove(self, padre_hi, padre_hd, actual, value): # método recursivo
    if actual == None:
      print("caso base")
      return None
    elif actual.data == value:
      print("\nEncontrado", actual.data)
      #caso 1: es nodo hoja
      if actual.left == None and actual.right == None:
        if padre_hi != None:
          padre_hi.left = None
        else:
          padre_hd.right = None
      #caso 2: tiene 1 hijo
      elif (actual.left != None and actual.right == None) or (actual.left == None and actual.right != None):
        if actual.left != None:
          if padre_hi!= None:
            padre_hi.left = actual.left
          else:
            padre_hd.right = actual.left
        else:
          if padre_hi!= None:
            padre_hi.left = actual.right
          else:
            padre_hd.right = actual.right               
      #caso 3: tiene los 2 hijos
    elif value < actual.data:
      self.__remove(actual, None, actual.left, value)
    else:
      self.__remove(None, actual, actual.right, value)
```

```
abb = ArbolBinarioBusqueda()
abb.insert(50)
abb.insert(30)
abb.insert(60)
abb.insert(35)
abb.insert(89)
abb.insert(55)
abb.insert(1)
abb.insert(33)
abb.insert(80)
abb.insert(100)
abb.insert(57)
abb.transversal()
abb.transversal("pre")
abb.transversal("pos")
res = abb.search(55)
print(f"El nodo es:{res.data}, hijo izquierdo es {res.left} e hijo derecho es {res.right}")
print(" ")
abb.remove(33)
abb.transversal()
abb.remove(55)
abb.transversal()
```
