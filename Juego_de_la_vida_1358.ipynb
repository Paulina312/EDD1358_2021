```
class Array2D:

  def __init__(self, rows, cols, value):
    self.__cols = cols
    self.__rows = rows
    self.__array = [[value for x in range (self.__cols)] for y in range(self.__rows)]
  
  def to_string(self):
    [print("---", end="") for x in range(self.__cols)]
    print("")
    for ren in self.__array:
      print(ren)
    [print("---", end="") for x in range(self.__cols)]
    print("")
  
  def get_num_rows(self):
    return self.__rows
  
  def get_num_cols(self):
    return self.__cols
  
  def get_item(self, row, col):
    try:
      return self.__array[row][col]
    except:
      return 0
  
  def set_item(self, row, col, valor):
    self.__array[row][col] = valor
  
  def clearing(self, valor = 0):
    for ren in range(self.__rows):
      for col in range(self.__cols):
        self.__array[ren][col] = valor

class JuegoDeLaVida:
  CELULA_VIVA = 1
  CELULA_MUERTA = 0
  
  def __init__(self, rens, cols, generaciones, poblacion):
    self.alto = rens
    self.largo = cols
    self.grid = Array2D(self.alto, self.largo, self.CELULA_MUERTA)
    self.grid.clearing(self.CELULA_MUERTA)
    self.generaciones = generaciones
    for celula in poblacion:
      self.grid.set_item(celula[0], celula[1], self.CELULA_VIVA)
  
  def imprime_grid(self):
    for r in range(self.grid.get_num_rows()):
      for c in range(self.grid.get_num_cols()):
        if self.grid.get_item(r,c) == 0:
          print("░░",end="")
        else:
          print("▓▓",end="")
      print("")
    print("\n")
  
  def configura_generacion(self, nueva_poblacion):
    self.grid.clearing()
    for celula in nueva_poblacion:
      self.grid.set_item(celula[0], celula[1], self.CELULA_VIVA)
  
  def get_num_rows(self):
    return self.alto
  
  def get_num_cols(self):
    return self.largo
  
  def set_celula_muerta(self, row, col):
    self.grid.set_item(row, col, self.CELULA_MUERTA)
  
  def set_celula_viva(self, row, col):
    self.grid.set_item(row, col, self.CELULA_VIVA)
  
  def get_is_viva(self, row, col): #<---?
    if self.grid.get_item(row, col) == self.CELULA_VIVA:
      return 1
    else:
      return 0
  
  def get_is_muerta(self, row, col):
    if self.grid.get_item(row, col) == self.CELULA_MUERTA:
      return 0
  
  def get_numero_vecinos_vivos(self, row, col):
    up_left = self.get_is_viva(row-1, col-1)
    up = self.get_is_viva(row-1, col)
    up_right = self.get_is_viva(row-1, col+1)
    left = self.get_is_viva(row, col-1)
    right = self.get_is_viva(row, col+1)
    down_left = self.get_is_viva(row+1, col-1)
    down = self.get_is_viva(row+1, col)
    down_right = self.get_is_viva(row+1, col+1)
    vecinos_vivos = (up_left + up + up_right + left + right + down_left + down + down_right)
    return vecinos_vivos
  
  def evolucionar(self):
    self.imprime_grid()
    i = 0
    while i <= self.generaciones:
      i = i+1
      print("Generación "+ str(i)+ "\n")
      nueva_poblacion = []
      nueva_poblacion.clear()
      for row in range(self.grid.get_num_rows()):
        for col in range(self.grid.get_num_cols()):
          if self.grid.get_item(row,col) == self.CELULA_MUERTA:
            elemento = self.get_numero_vecinos_vivos(row,col)
            if elemento == 3:
              nueva_poblacion.append((row, col))
            else:
              continue
          elif self.grid.get_item(row, col) == self.CELULA_VIVA:
            elemento = self.get_numero_vecinos_vivos(row,col)
            if elemento == 2:
              nueva_poblacion.append((row, col))
            elif elemento == 3:
             nueva_poblacion.append((row, col))
            else:
              continue
      self.configura_generacion(nueva_poblacion)
      self.imprime_grid()

experimento = JuegoDeLaVida(7, 7, 5, [(1,2), (2,1), (2,2), (2,3)])
experimento.evolucionar()
```
